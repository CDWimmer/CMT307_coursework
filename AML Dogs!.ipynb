```python
# Set up some blank arrays to hold data
image_size = []  # to be filled with a tuple (x,y) for each image

image_bbox_size = [] # to be filled with a tuple (x,y) for each bndbox


image_depth = []  # I believe this refers to colour channels? Should be 3 

breed_count = dict()

total_images = 0
total_dogs = 0
```

```python
# define function for processing meta_dict["object"] data
def object_processor(meta_dict_obj):
    global total_dogs
    total_dogs += 1
    image_bbox_size.append(
            (int(meta_dict_obj["bndbox"]["xmax"]) - int(meta_dict_obj["bndbox"]["xmin"]),
            int(meta_dict_obj["bndbox"]["ymax"]) - int(meta_dict_obj["bndbox"]["ymin"]))
    )
    breed = meta_dict_obj["name"]
    if breed not in breed_count:
        breed_count[breed] = 1
    else:
        breed_count[breed] += 1  
```

```python
print("This might take a few seconds...")
# open all the files and populate lists with the data
for metadata in mat["annotation_list"]:
    total_images += 1
    with open(f"{annotation_path}/{metadata[0][0]}", 'rb') as f:
        meta_dict = xmltodict.parse(f.read())["annotation"]
        image_size.append( (int(meta_dict["size"]["width"]), int(meta_dict["size"]["height"])) )
        
        if isinstance(meta_dict["object"], list):  # if there is more than one dog then 
            for obj_dict in meta_dict["object"]:   # there is more than one object
                object_processor(obj_dict)
        else:  # when there's only one dog:
            object_processor(meta_dict["object"])

        image_depth.append(int(meta_dict["size"]["depth"]))

print("Done!")
```

### We can now do some further calculations

```python
image_area = [int(axis[0])*int(axis[1]) for axis in image_size]
bndbox_area = [int(axis[0])*int(axis[1]) for axis in image_bbox_size]
avg_area = np.mean(image_area)
avg_bbox = np.mean(bndbox_area)

pixel_reduction = (avg_bbox / avg_area) * 100  # percentage

avg_depth = np.mean(np.array([image_depth]).astype(np.float))  # I don't know why I had to do this, this isn't even useful

breed_table = pd.DataFrame.from_dict(breed_count, orient='index', columns=["Count"]).reset_index()
breed_table.columns = ["Breed", "Count"]
```

```python
print(f"A total of {total_dogs} dogs from {total_images} images")
print(f"Average area of original images: {avg_area:.3f} px") 
print(f"Average area of bounding boxes: {avg_bbox:.3f} px") 
print(f"Approximate overal reduction in size: {pixel_reduction:.3f}%")

print("Breed information:")
breed_table
```
